```python
from urllib.request import urlopen
from bs4 import BeautifulSoup

page =  urlopen("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M").read()
soup =  BeautifulSoup(page)


data = []
table = soup.find("table", { "class" : "wikitable sortable" })
table_body = table.find('tbody')

rows = table_body.find_all('tr')
for row in rows:
    cols = row.find_all('td')
    cols = [ele.text.strip() for ele in cols]
    data.append([ele for ele in cols if ele]) # Get rid of empty values

data
```

```python
import pandas as pd
df = pd.DataFrame(data, columns=["Postal Code", "Borough","Neighborhood"]) 

df
```

```python
df = df[~df['Neighborhood'].isin(['Not assigned'])]
df
```

```python
df1 = df.groupby(['Postal Code', 'Borough'])['Neighborhood'].apply(','.join).reset_index()
df1.columns = ['Postal Code', 'Borough', 'Neighborhood']
df1
```

```python
df1.loc[df1['Neighborhood']=="Not assigned",'Neighborhood']=df1.loc[df1['Neighborhood']=="Not assigned",'Borough']
df1
```
